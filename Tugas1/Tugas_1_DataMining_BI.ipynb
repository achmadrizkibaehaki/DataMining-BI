```python
Jakarta_created_at_column = df['Jakarta']['created_at']

# Initialize an empty set to store unique date formats
unique_formats = set()

# Iterate through each value in the 'created_at' column
for date_str in Jakarta_created_at_column:
    # Try to parse the date using different formats
    for date_format in ['%Y-%m-%d', '%d-%m-%Y', '%m/%d/%Y', '%d/%m/%Y', '%Y%m%d']:
        try:
            parsed_date = pd.to_datetime(date_str, format=date_format)
            # If parsing is successful, add the format to the set
            unique_formats.add(date_format)
            break  # Break out of the inner loop if successful to move to the next value
        except ValueError:
            pass  # Continue to the next format if parsing fails

# Print the unique date formats
print("Unique date formats in 'created_at':", unique_formats)
```

```python
merged_df['listrik'].unique()
```

diketahui beberapa input data listrik ada yang memiliki kata watt, Va/R-1, titik. agar lebih mudah divisualisasi akan diubah agar hanya keliatan angkanya

```python
# Assuming merged_df is your DataFrame and 'listrik' is the column you want to modify
merged_df['listrik'] = merged_df['listrik'].replace({
    r'\.': '',                 # Remove dots
    r'\s*Watt$': '',           # Remove 'Watt' at the end
    r'\s*KWh$': '000',         # Replace 'KWh' with '000'
    r'[^0-9]': '',             # Remove non-numeric characters
}, regex=True)

# Convert the 'listrik' column to numeric and then to integer
merged_df['listrik'] = pd.to_numeric(merged_df['listrik'], errors='coerce').astype('Int64')

# Replace '-' and NaN with pd.NA
merged_df['listrik'] = merged_df['listrik'].replace({'-': pd.NA, pd.NA: pd.NA})

# Change the data type of the 'listrik' column to object (string)
merged_df['listrik'] = merged_df['listrik'].astype(str)

# Fill remaining missing values with "Tidak diketahui"
merged_df['listrik'] = merged_df['listrik'].fillna('Tidak diketahui')

# Display unique values in the 'listrik' column after cleaning
print("Unique values in the 'listrik' column after cleaning:")
print(merged_df['listrik'].unique())
```

```python
import numpy as np

# Assuming merged_df is your DataFrame and 'hadap' is the column you want to modify
merged_df['hadap'] = merged_df['hadap'].str.lower().str.strip()  # Convert to lowercase and remove leading/trailing whitespaces

# Map specific directions to broader categories
direction_mapping = {
    'barat': 'Barat',
    'selatan': 'Selatan',
    'timur': 'Timur',
    'utara': 'Utara',
    'danau': 'Danau',
    'tenggara': 'Tenggara',
    'barat daya': 'Barat Daya',
    'timur laut': 'Timur Laut',
    'barat laut': 'Barat Laut',
}

# Replace specific directions with broader categories
merged_df['hadap'] = merged_df['hadap'].replace(direction_mapping)

# Replace more than two directions with 'Dua Arah'
multi_directions = [
    'selatan dan barat', 'selatan dan utara', 'selatan & barat',
    'barat & selatan', 'selatan dan timur', 'timur - selatan',
    'utara - barat', 'utara - selatan', 'selatan - barat', 'utara-timur',
    'timur-utara', 'utara-barat', 'utara dan timur (hook)', 'selatan/timur'
]

merged_df['hadap'] = merged_df['hadap'].replace(multi_directions, 'Dua Arah')

# Replace '-' and NaN with "Tidak Diketahui"
merged_df['hadap'] = merged_df['hadap'].replace({'-': 'Tidak Diketahui', np.nan: 'Tidak Diketahui'})

# Capitalize the first letter for specific cases
merged_df['hadap'] = merged_df['hadap'].replace({'timur utara': 'Timur Utara', 'barat': 'Barat'})

# Display unique values in the 'hadap' column after cleaning
print("Unique values in the 'hadap' column after cleaning:")
print(merged_df['hadap'].unique())
```
